```python
class color:
    PURPLE = '\033[95m'
    CYAN = '\033[96m'
    DARKCYAN = '\033[36m'
    BLUE = '\033[94m'
    GREEN = '\033[92m'
    YELLOW = '\033[93m'
    RED = '\033[91m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    END = '\033[0m'
```

```python
from firebase_admin import db
print('Drinks')
print()
ref = db.reference('Drinks')
snapshot = ref.order_by_key().get()
print(color.BOLD +  "{:<18} {:<15} {:<13} {:<20}".format('Product name','MRP(in Rs.)','MFD', 'Quantity left') + color.END)
print()
for key, val in snapshot.items():
    print("{:<21} {:<12} {:<15} {:<20}".format(key, val['MRP'], val['MFD'], val['Quantity left']))
#     print(key)
#     print('MFD', ':', val['MFD'], '  ', end = '')
#     print('MRP', ':', val['MRP'], ':  ', end = '')
#     print('Quantity left', ':', val['Quantity left'], ':   ')
```

```python
import math
import random
d = {1: ["Spices", math.floor(random.gauss(40, 5))],
    2: ["Other stuff", math.floor(random.gauss(20, 5))],
    3: ["Tea", math.floor(random.gauss(50, 5))],
    10: ["Contraband", math.floor(random.gauss(1000, 5))],
    5: ["Fruit", math.floor(random.gauss(10, 5))],
    6: ["Textiles", math.floor(random.gauss(40, 5))]
}
print( "{:<8} {:<15} {:<10}".format('Key','Label','Number'))
for k, v in d.items():
    label, num = v
    print("{:<8} {:<15} {:<10}".format(k, label, num))
```
